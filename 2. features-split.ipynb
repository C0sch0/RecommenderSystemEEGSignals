```python
import numpy as np
import json
```

```python
ROOT_PATH = "/content/drive/Shareddrives/Proyecto RecSys 2021-2/Proyecto"
FEATURES_PATH = f"{ROOT_PATH}/new_features"
SPLITTED_FEATURES_PATH = f"{ROOT_PATH}/features-and-labels"
LABELS_PATH = f"{ROOT_PATH}/new_labels"
```

```python
NUM_USERS = 32
NUM_VIDEOS = 40
NUM_FRAMES = 29
FEATURES_DIM = 1024
```

```python
TEST_SIZE = 10

np.random.seed(0)

idxs = list(range(NUM_VIDEOS))

TEST_VIDEO_IDXS = np.random.choice(idxs, TEST_SIZE, replace=False)
TRAIN_VIDEO_IDXS = np.array([c for c in range(NUM_VIDEOS) if c not in TEST_VIDEO_IDXS])

# print(TEST_VIDEO_IDXS)
# print(TRAIN_VIDEO_IDXS)
```

```python
user_video_features = lambda user, video: f"{FEATURES_PATH}/usuario_{str(user + 1).zfill(2)}/video_{video}/features.npy"
```

---

```python
user_features = np.zeros((NUM_USERS, NUM_VIDEOS, NUM_FRAMES, FEATURES_DIM))

for user in range(NUM_USERS):
    video_features = np.zeros((NUM_VIDEOS, NUM_FRAMES, FEATURES_DIM))

    for video in range(NUM_VIDEOS):
        print(f"User {user}, video {video}: ", end='')

        frame_features = np.zeros((NUM_FRAMES, FEATURES_DIM))

        with open(user_video_features(user, video), 'rb') as file:
            for frame in range(NUM_FRAMES):
                frame_features[frame] = np.squeeze(np.load(file))

                print(frame, end=' ')
        video_features[video] = frame_features

        print()
    user_features[user] = video_features
```

```python
user_features.shape
```

```python
with open(f"{SPLITTED_FEATURES_PATH}/features_all.npy", 'wb') as file:
    np.save(file, user_features)
```

---

```python
labels = np.zeros((NUM_USERS, NUM_VIDEOS), dtype=np.int8)

for user in range(NUM_USERS):
    with open(f"{LABELS_PATH}/user_{user}", "rb") as file:
        labels[user] = np.load(file)
```

```python
with open(f"{SPLITTED_FEATURES_PATH}/labels_all.npy", 'wb') as file:
    np.save(file, labels)
```

---

```python
with open(f"{SPLITTED_FEATURES_PATH}/labels_all.npy", 'rb') as file:
    labels = np.load(file)

with open(f"{SPLITTED_FEATURES_PATH}/features_all.npy", 'rb') as file:
    features = np.load(file)

train_labels = labels[:, [TRAIN_VIDEO_IDXS]]
test_labels = labels[:, [TEST_VIDEO_IDXS]]

train_features = features[:, [TRAIN_VIDEO_IDXS], :, :]
test_features = features[:, [TEST_VIDEO_IDXS], :, :]
```

```python
with open(f"{SPLITTED_FEATURES_PATH}/train_labels.npy", 'wb') as file:
    np.save(file, train_labels)

with open(f"{SPLITTED_FEATURES_PATH}/test_labels.npy", 'wb') as file:
    np.save(file, test_labels)

with open(f"{SPLITTED_FEATURES_PATH}/train_features.npy", 'wb') as file:
    np.save(file, train_features)

with open(f"{SPLITTED_FEATURES_PATH}/test_features.npy", 'wb') as file:
    np.save(file, test_features)
```
