```python
import matplotlib.pyplot as plt
import numpy as np
import mne
```

```python
def load_raw_data(path):
    with open(file_path, 'rb') as file:
        raw = pickle.load(file, encoding='latin-1')

    return raw["data"], raw["labels"]
```

```python
def filter_channels(data):
    return data[:, :32, :]
```

```python
SAMPLE_RATE = 128.  #freq de medida del EEG
NUM_CHANNELS = 32
```

```python
def generate_eeg_image(data_sample, freq):
    biosemi_montage = mne.channels.make_standard_montage('biosemi32')
    channel_names = biosemi_montage.ch_names

    info = mne.create_info(
        ch_names=channel_names,
        sfreq=SAMPLE_RATE,
        ch_types='eeg'
    )

    evoked_array = mne.EvokedArray(data_sample, info)
    evoked_array.set_montage(biosemi_montage)
    
    return mne.viz.plot_topomap(
        evoked_array.data[:, freq], 
        evoked_array.info, 
        cmap='nipy_spectral', 
        outlines=None,
        show=False
    )
```

```python
file_nums = [str(i).zfill(2) for i in range(1, 33)]
seconds = np.arange(0, 8064, 134)
```

```python
for file_num in file_nums:
  file_path = f"{root_path}/databases/DEAP/data_preprocessed_python/s{file_num}.dat"
  raw_data, labels = load_raw_data(file_path)
  filtered_channels_data = filter_channels(raw_data)
  for k in range(41):
    visualization_data = filtered_channels_data[k]

    for second in seconds:
      fig = plt.figure(figsize=(3,3))
      m = generate_eeg_image(visualization_data, second)
      plt.savefig(f'{root_path}/images/test-image.png')  # changed for safety
      plt.close(fig)

      break
    break
  break
```
